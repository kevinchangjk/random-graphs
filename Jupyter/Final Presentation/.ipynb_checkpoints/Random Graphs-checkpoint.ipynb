## What are Random Graphs?

Random graphs are graphs of nodes or vertices, which are connected by edges. These connections and nodes are generated either by a random distribution, or a random process. 

There are many different models that use different methods to create random graphs, to obtain different graphs with different properties.

## What properties do we want?

Different models are suited for different applications. 

In our case, we focus on modelling social networks, thus we will be discussing the common characteristics that these networks display.

## Clustering

A commonly studied property of random graphs is the clustering coefficient. 

The local clustering coefficient is the probability that, if you know two friends, the two of them would know each other as well.

This is the clustering coefficient for a specific node $i$, $C_i$, and is calculated as the no. of existing edges between neighbouring nodes divided by the no. of potential edges between neighbouring nodes.



## The Big Picture

Having defined the local clustering coefficient for a single node, we can apply this to the whole graph to get a macro view of its clustering.

One way to do this is by taking the mean of all the local clustering coefficients. Thus we arrive at $C_g = \sum \frac {C_i} {n}$

This is the average clustering coefficient.

## Average vs Global

Another commonly used measure for clustering is transitivity, or the global clustering coefficient. This is a weighted average of the local clustering coefficient, where each $C_i$ is multiplied by ${k_i} \choose {2}$, where $k_i$ is the degree of the node $i$, the number of nodes it is connected to.

Transitivity places more weight on higher degree nodes, which are commonly called "hubs", while the average clustering coefficient places equal weight on all nodes.


## It's a Small World

Researchers often say that we live in a small world. This means that we can reach out to almost anyone in the world with just a few connections. This theory is popularised as the six degrees of separation. 

Small world behaviour in networks is characterised largely by high clustering and a low mean path distance. The path distance between two nodes is the smallest number of edges required to connect the two nodes.


## Degree Distributions

The degree distribution of a graph is a histogram of the degrees of all the nodes in the graph.


## Scale-free degree distribution

For a social network, a scale-free degree distribution is a key property. This means that the degree distribution follows a power law, and is skewed such that there is a very low number of hubs, and a much higher number of nodes with degree close to zero.

## Watts-Strogatz Model

Also called the small world model, this is a model meant to capture the key elements of a small world network. 

The Watts-Strogatz Model requires 3 parameters, $n$, $k$, and $\beta$. 

A graph of $n$ nodes is first created, with each node connected to $k/2$ nodes to their immediate left and right. Each edge in the graph is then rewired with probability $\beta$ to a random node.

As such, if $\beta$ is small, the graph forms a structure similar to a lattice, and a few rewired edges are enough to significantly reduce the mean path distance.

```python
def watts_strogatz(n, k, beta):
    
    def node_gen(node):
        interval = 2*np.pi / n
        norm, theta = [1, interval * node]
        x, y = np.cos(theta) * norm, np.sin(theta) * norm
        return (node, dict(pos = tuple([x, y])))
    
    nodes = list(map(node_gen, range(n)))
    g = nx.Graph()
    g.add_nodes_from(nodes)

    edges = []
    def edges_add(node):
        targets = list(range(n))
        targets.remove(node)
        b = np.random.ranf(k//2) < beta
        addition = [(node, node+i) if node+i < n else (node, node+i-n) for i in range(1, k//2+1)]
        addition = [addition[i] if b[i] == 0 else (node, np.random.choice(targets)) for i in range(k//2)]
        edges.extend(addition)
        return
    edge = list(map(edges_add, range(n)))
    g.add_edges_from(edges)
    return g
```

## Barabási-Albert Model

The Barabási-Albert model uses two parameters, $n$ for the number of nodes, and $m$ for the number of edges per node. The model begins with $m$ nodes in an empty graph, and at each next step, a node is added, with m edges connected to other nodes. 

The probability of the new node being connected to a node is linearly proportional to the degree of the node. This method is known as linear preferential attachment.

```python
def barabasi_albert(n, m):
    
    def node_gen(node):
        interval = 2*np.pi / n
        norm, theta = [1, interval * node]
        x, y = np.cos(theta) * norm, np.sin(theta) * norm
        return (node, dict(pos = tuple([x, y])))
    
    nodes = list(map(node_gen, range(n)))
    g = nx.Graph()
    g.add_nodes_from(nodes)
    
    edges_initial = []
    for node in range(m):
        edges_initial.extend([(node, target) for target in range(node+1, m)])
        
    g.add_edges_from(edges_initial)
    
    def edge_add(node):
        available = list(range(n))
        available.remove(node)
        
        def pcalc(available):
            return np.array([g.degree(ni) for ni in available])
        
        p_dist = pcalc(available)
        p_dist = p_dist / sum(p_dist)
        
        targets = np.random.choice(available, size = m, p = p_dist)
        return [(node, target) for target in targets]
    
    for node in range(m, n):
        g.add_edges_from(edge_add(node))
        
    node_sizes = [g.degree(node) * 10 for node in range(n)]
    
    return g, node_sizes
```

## Klemm-Eguíluz

This is a small world and scale-free model that is a crossover of a high clustering model and the Barabasi-Albert model. Three parameters are required, $n$, $m$, and $\mu$.

1. It begins with a connected graph of $m$ number of active nodes, and a new node is added in each time step. 

2. The new node has $m$ edges added to the $m$ active nodes, and each edge is decided, with probability $\mu$, that the edge is rewired to a random node according to linear preferential attachment. 

3. The new node is then active, and a random node is deactivated, such that $m$ active nodes remain.

Thus, depending on $\mu$, the graph can be more of a high clustering model, or more like the Barabasi-Albert Model, essentially combining both models.

```python
def klemm_eguiluz(n, m, mu):
    
    def node_gen(node):
        interval = 2*np.pi / n
        norm, theta = [1, interval * node]
        x, y = np.cos(theta) * norm, np.sin(theta) * norm
        return (node, dict(pos = tuple([x, y])))
    
    nodes = list(map(node_gen, range(n)))
    g = nx.Graph()
    g.add_nodes_from(nodes)
    
    edges_initial = []
    for node in range(m):
        edges_initial.extend([(node, target) for target in range(node+1, m)])
        
    g.add_edges_from(edges_initial)
    
    activation = list(np.ones(m))
    mu_factor = np.random.ranf(m)
    
    for i in range(m, n):
        # generate list of edges which are to be randomly rewired
        mu_factor = np.random.ranf(m) < mu
        activated = np.where(np.array(activation) == 1)[0]

        targets = set([activated[j] for j in range(m) if mu_factor[j] == 0])
        
        # Linear preferential attachment
        p_total = np.array(list(map(g.degree, range(i))))
        p_dstr = p_total / sum(p_total)
        
        while len(targets) < m:
            targets.add(np.random.choice(list(range(i)), p = p_dstr))
        g.add_edges_from(zip(np.full(m, i), list(targets)))
        
        # Activation and deactivation where p = a / k, and 1/a = sum of 1/k
        
        k =  np.array([g.degree(active)**-1 for active in activated])
        a = sum(k)
        p_deact = k / a
        deactivated = np.random.choice(activated, p = p_deact)
        activation[deactivated] = 0
        activation.append(1)
    
    node_sizes = [g.degree(node) * 1000/n for node in range(n)]
    
    return [g, node_sizes]
```

### SIR

We might not be able to find perfect model for our social network yet, but let us see if they can act in a similar way to Facebook.

In social networks, data and information is constantly exchanged. People spread information to their friends, family, who spread it on to their friends and family, and so on. 


To analyse this, we have the Susceptible-Infected-Recovered (SIR) Model. 

Nodes are in one of three states: susceptible, infected, or recovered. All nodes begin in the susceptible state. A few are infected, then at every time step, the neighbours of the infected nodes are infected with probability $p_1$, and the infected nodes recover with probability $p_2$. Once all the nodes are recovered, the model ends.

While the most common use of this is to model the spread of infectious diseases, we will apply this to our random graphs to analyse the spread of information, and the properties of random graphs that are more significant here.

```python
def sir_plot(g, si, ir, ini, tn):

    size = len(g.nodes())
    
    inin = np.random.ranf(size) < ini
    state = inin * 1
    i_plot = [sum(state)]
    r_plot = [0]
    s_plot = [size-i_plot[0]]
    
    def infect():
        base = nx.adjacency_matrix(g)
        infected = state == 1
        infected_p = base * infected
        infected_pdist = np.random.ranf(size) < 1-(1-si)**infected_p
        infected_new = [i if state[i] == 0 and infected_pdist[i] == 1 else -1 for i in range(size)]
        infected_new = np.array(list(filter(lambda x: x >=0, infected_new)))
 
        state_new = state.copy()
        if np.size(infected_new) > 0:
            state_new[infected_new] = 1
        recovery = np.random.ranf(size) < ir
        recovered = [i if recovery[i] == 1 and state[i] == 1 else -1 for i in range(size)]
        recovered = np.array(list(filter(lambda x: x >= 0, recovered)))
        if np.size(recovered) > 0:
            state_new[recovered] = -1
        
        return state_new
    
    for t in range(1, tn+1):
        state = infect()
        state_count = Counter(state)
        s_plot.append(state_count[0])
        i_plot.append(state_count[1])
        r_plot.append(state_count[-1])

    plt.plot(range(0, t+1), s_plot, color = 'b', label = 'Susceptible')
    plt.plot(range(0, t+1), i_plot, color = 'r', label = 'Infected')
    plt.plot(range(0, t+1), r_plot, color = 'g', label = 'Recovered')
    plt.legend(loc = 'upper right')
    plt.xlabel('Time')
    plt.ylabel('Number of nodes')
    plt.show()
    return 

def sir_init(g, si, ir, ini):
    size = len(g.nodes())
    
    def color(n):
        if state[n] == 0:
            return 'b'
        elif state[n] == 1:
            return 'r'
        else:
            return 'g'
    
    inin = np.random.ranf(size) < ini
    state = inin * 1
    i_plot = [sum(state)]
    r_plot = [0]
    s_plot = [size-i_plot[0]]
    
    t = 0
    colors = list(map(color, range(size)))
    print(f"\nSIR model of graph after {t} time steps:")
    nx.draw_shell(g, node_color = colors)
    plt.show()
    
    return [i_plot, r_plot, s_plot, state, t]

def sir_step(state, t):
    size = len(g.nodes())
    si = p1
    ir = p2
    
    def color(n):
        if state[n] == 0:
            return 'b'
        elif state[n] == 1:
            return 'r'
        else:
            return 'g'
    
    def infect():
        base = nx.adjacency_matrix(g)
        infected = state == 1
        infected_p = base * infected
        infected_pdist = np.random.ranf(size) < 1-(1-si)**infected_p
        infected_new = [i if state[i] == 0 and infected_pdist[i] == 1 else -1 for i in range(size)]
        infected_new = np.array(list(filter(lambda x: x >=0, infected_new)))
 
        state_new = state.copy()
        if np.size(infected_new) > 0:
            state_new[infected_new] = 1
        recovery = np.random.ranf(size) < ir
        recovered = [i if recovery[i] == 1 and state[i] == 1 else -1 for i in range(size)]
        recovered = np.array(list(filter(lambda x: x >= 0, recovered)))
        if np.size(recovered) > 0:
            state_new[recovered] = -1
        
        return state_new
    
    state = infect()
    state_count = Counter(state)
    s_plot.append(state_count[0])
    i_plot.append(state_count[1])
    r_plot.append(state_count[-1])
    t+=1
        
    colors = list(map(color, range(size)))
    print(f"\nSIR model of graph after {t} time steps:")
    nx.draw_shell(g, node_color = colors)
    plt.show()
    
    return [state, t]
```

## Stochastic Block Model

Stochastic block models are random graphs formed by first dividing the available nodes into groups, then connecting the nodes using different values of $p_1$ and $p_2$ to increase the connections within groups, and to sparsely connect nodes in different groups.

As a result, blocks or clusters are formed, with few connections to other clusters.

```python
def block(n, p1, p2, zeta = 0.3):
    groups = []

    while sum(groups) < n:
        ppln = np.random.randint(1, n * zeta)
        if ppln + sum(groups) > n:
            ppln = n - sum(groups)
        groups.append(ppln)
        
    edges = np.random.ranf([n, n])
    edges = (edges + edges.T) / 2
    targets = edges.copy()
    targets = targets < (p2*2)
    group = 0
    node = 0
    while group < len(groups):
        append = edges[node:node+groups[group], node:node+groups[group]] < p1 * 2
        targets[node:node+groups[group], node:node+groups[group]] = append
        node = node + groups[group]
        group +=1
    
    print(f"Number of clusters: {len(groups)}")
    g = nx.from_numpy_matrix(targets)
    return g
```

## Geometric Graphs

Geometric random graphs are formed by a different process compared to the other models.

The model begins by placing nodes over the graph. Then all node-pairs that have an Euclidean distance less than a certain critical value would be connected with an edge. 

As such, nodes would be connected to every other node within a certain radius, and no other nodes.

```python
def post_gen(t):
    theta, norm = np.random.ranf() * np.pi * 2, np.random.ranf()
    x, y = np.cos(theta) * norm, np.sin(theta) * norm
    return [x, y]

def geometric_graph(n, d):
    g = nx.Graph()
    posts = list(map(post_gen, range(n)))
    
    def node_add(node):
        return (node, dict(pos = posts[node]))
    
    nodes = list(map(node_add, range(n)))
    g.add_nodes_from(nodes)
    
    def edges_add(node):
    
        def distance(target):
            dist = ((posts[node][0] - posts[target][0]) ** 2 + (posts[node][1] - posts[target][1]) ** 2)**0.5
            return dist
        
        distance = np.vectorize(distance)
        dists = distance(range(n))
        targets = np.where(dists < d)[0]
        return zip(targets, np.full(len(targets), node))
    
    edges_add = np.vectorize(edges_add)
    edges = edges_add(range(n))
    for edge in edges:
        g.add_edges_from(edge)
        
    return g
```

```python
n = 100
d = 0.4

g = geometric_graph(n, d)
pos = nx.get_node_attributes(g, 'pos')
nx.draw(g, pos, node_size = 150)
```

## Chang-Chang Model

A model inspired from block models and geometric graphs. It has five parameters, $n$, $k$, $\chi$, $\phi$, $\delta$.

The graph is created by first adding $\chi$ nodes with random coordinates. $\chi$ is the number of clusters and these nodes are the cluster 'bases' that other nodes will be concentrated around. 

At every time step, a new node is added. This new node is grouped into one of the $\chi$ clusters, and is added with random coordinates scattered around the cluster base.

A number of edges are then added to the node. These edges will be connected to other nodes as well, and the probability of it connecting to a node is inversely proportional to the Euclidean distance between the two points.

```python
def post_gen():
    theta, norm = np.random.ranf() * np.pi * 2, np.random.ranf()
    x, y = np.cos(theta) * norm, np.sin(theta) * norm
    return [x, y]

def chang_chang(n, k, phi, chi, delta):
    
    g = nx.Graph()
    
    # coordinates for cluster bases
    clusters = [[int(i)] for i in range(chi)]
    cluster_b = [tuple(post_gen()) for i in clusters]
    posts = cluster_b.copy()
    
    # function to return formatted tuple to add position with node
    def cluster_b_gen(node):
        return (node, dict(pos = cluster_b[node]))
    
    initial = list(map(cluster_b_gen, range(chi)))
    g.add_nodes_from(initial)
    
    # function to add chi edges among first chi nodes; ensure graph is connected. Not implemented currently
    def edge_initialize(node):
        target = np.random.choice(range(chi))
        return (node, target)
    
    edges_initialize = list(map(edge_initialize, range(chi)))
    g.add_edges_from(edges_initialize)
    
    for node in range(chi, n):
        # Adding node into one of the clusters. Not implementing preferential addition now.
        '''
        pplns = list(map(len, clusters))
        ppln = list(np.array(pplns) / sum(pplns))
        '''
        cluster_add = np.random.choice(range(chi))
        i = cluster_add
        clusters[i].append(node)
        
        # Scaling random coordinates generated with post_gen by delta, returning new coordinates for point
        def delt_gen(node):
            delt = post_gen()
            delt = tuple([delt[0] / delta, delt[1] / delta])
            post = (cluster_b[i][0] + delt[0], cluster_b[i][1] + delt[1])
            posts.append(post)
            return post
        
        g.add_node(node, pos = delt_gen(node))

        def edges(k):
            
            # Calculating distance between points
            def distance(target):
                dist = ((posts[node][0] - posts[target][0]) ** 2 + (posts[node][1] - posts[target][1]) ** 2)**0.5
                return dist
            
            # Probability distribution where p = (k+1)^phi / d, k+1 is used so nodes with zero degree can be chosen
            def pcalc(target):
                d = distance(target)
                return ((g.degree(target)+1)**phi) / d
            
            def p_dist(targets):
                dist = np.array(list(map(pcalc, targets)))
                dist = dist / sum(dist)
                return list(dist)
            
            # determining edges and targets of edges
            size_n = np.random.randint(1, k)
            targets = np.random.choice(range(0, node), p = p_dist(range(0, node)), size = size_n, replace = False)
            edges_add = list(zip(targets, np.full(len(targets), node)))
        
            return edges_add
    
        g.add_edges_from(edges(k))
        
    node_sizes = [(g.degree(node) * 10)**0.5 for node in range(n)]
    
    return [g, node_sizes]
```

```python
n = 4039
k = 44
phi = 2
chi = 400
delta = 30
g, node_sizes = chang_chang(n, k, phi, chi, delta)
```

## Chang-Chang Characteristics

The proposed model has high average clustering coefficient, but a low transitivity.

Recalling that transitivity is the weighted average of the local clustering coefficients by ${k}\choose{2}$, this suggests that the clustering coefficient of lower degree nodes is much higher than those with higher degrees, the hubs. 

The shape of the graph in general, is much messier than Facebook, and indicates that there are too many connections between the clusters.

The degree distribution of the graph is scale-free, and follows a power law. The mean path distance is also low at 2.203.

Overall, while the model may have some features that match Facebook's, it is not an accurate model.

## Summary

Random graphs can be used to model social networks as one of its applications.

Social networks would have a few main properties that would be desirable in random graphs. These are mainly a high amount of clustering, a scale-free degree distribution, a low mean path distance.
